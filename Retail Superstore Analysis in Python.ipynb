### **Data Cleaning**

```python
print(df['Category'].unique())
```

```python
df['Category']= df['Category'].replace({'furniture':'Furniture',
                                            'F':'Furniture',
                                            'office Supplies':'Office Supplies',
                                            'OS':'Office Supplies',
                                            'technology':'Technology'
                                           })
                                            
```

### **Business Requirements**

### **KPI's Requirements**

```python
#Total sales
total_sales = df['Sales'].sum()

#Average Sales
avg_sales = df['Sales'].mean()

#No Of Items Sold
no_of_items_sold = df['Sub-Category'].count()

#Average Ratings
avg_ratings = df['Rating'].mean()

#Display

print(f"Total sales: ${total_sales:,.1f}")
print(f"Average sales: ${avg_sales:,.1f}")
print(f"No Of Items Sold: {no_of_items_sold:,.0f}")
print(f"Average Ratings: {avg_ratings:,.1f}")
```

### **Chart's Requirements**

#### **Total Sales by Category**

```python
sales_by_category = df.groupby('Category')['Sales'].sum()
plt.pie(sales_by_category, labels = sales_by_category.index,
                           autopct = '%.1f%%',
                           startangle = 90)
plt.title(' Sales by category')
plt.axis('equal')
plt.show()
```

#### **Total Sales by Sub-Category**

```python
sales_by_sub_category = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,8))
bars = plt.bar(sales_by_sub_category.index, sales_by_sub_category.values)

plt.xticks(rotation=90)
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Sub-Category')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

#### **Category by Region for Total Sales**

```python
grouped = df.groupby(['Region','Category'])['Sales'].sum().unstack()
grouped = grouped[['Furniture', 'Office Supplies', 'Technology']]

ax = grouped.plot(kind='bar',figsize=(10,6),title='Category by Region Sales')
plt.xlabel('Category by Region')
plt.ylabel('Total Sales')
plt.legend(title='Category by Region')
plt.tight_layout()
plt.show()
```

#### **Total Profit by Discount**

```python
sales_by_discount = df.groupby('Discount')['Profit'].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(sales_by_discount.index, sales_by_discount.values, marker='o', linestyle='-')

plt.xlabel('Discount')
plt.ylabel('Total Profit')
plt.title('Profit by Discount')

for x,y in zip(sales_by_discount.index, sales_by_discount.values):
    plt.text(x,y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

#### **Top 5 Rating by Sub-Category**

```python
sales_by_rating = df.groupby('Sub-Category')['Rating'].sum().tail(5)

plt.figure(figsize=(7,7))
plt.pie(sales_by_rating, labels=sales_by_rating.index, autopct='%1.1f%%', startangle=90)
plt.title('Rating by Sub-Category')
plt.tight_layout()
plt.show()
```

#### **Sales by Segment**

```python
sales_by_segment = df.groupby('Segment')['Sales'].sum().reset_index()
sales_by_segment = sales_by_segment.sort_values('Sales', ascending=False)

plt.figure(figsize=(8,4))
ax = sns.barplot(x='Sales', y='Segment', data=sales_by_segment)

plt.xlabel('Total sales')
plt.ylabel('Segmnet')
plt.title('Sales by Segment')

plt.tight_layout()
plt.show()
```
